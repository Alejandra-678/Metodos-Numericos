## **3.3 Diccionarios**

Permiten utilizar una clave para declarar y acceder a un valor

$$ \text{Diccionario}= \text{{"Clave1": Valor1, "Clave2": Valor2,...} }$$

````python
#Declarar diccionario
Dic = {
    "Día": 6, 
    "Mes": "Julio",
    "Año": 2021,
    "Paciente":"Alain Espinoza",
    "Médico": "Dra. Laura Steinwender",
    "Consultorio": 41
}
````

````python
#Acceder a un elemento del diccionario (2 métodos sencillo y .get())
print(Dic["Médico"])
print(Dic.get("Médico"))
````

````python
# Items Primero será la clave y el segundo dic.items()
Dic.items()
````

````python
# Keys retorna claves dic.keys()
Dic.keys()
````

````python
# Values retorna valores dic.values()
Dic.values()
````

````python
#Zip Recibe como parámetro dos elementos iterables dic = dict(zip('abcd',[1,2,3,4]))
dic = dict(zip('abcd',[1,2,3,4]))
print(dic)
````

# **5. Ciclos y Condicionales**

## **5.1 If, elif, else**

Los condicionales los utilizaremos para validar si se cumple una o más condiciones para ejecutar cierto bloque de código, en esencia nos ayuda a tomar decisiones sobre el flujo de nuestro código.

```python
if condition:
    # Block of code
elif other_condition:
	# Another block of code
else:
	# Final block of code
```

alertas.PNG

````python
semaforo="rojo"

if semaforo == "verde":
    print("Nueva Normalidad")
elif semaforo=="amarillo":
	  print("Alerta Moderada")
elif semaforo=="anaranjado":
	  print("Alerta Alta")
else:
	  print("Alerta Máxima")
````

````python
compra=500
if compra <= 100:
    print("Pago en efectivo")
elif compra > 100 and compra < 300:
    print("Pago con tarjeta de débito")
else:
    print("Pago con tarjeta de crédito")
````

## **5.2 For**

Python tiene un ciclo basado en colecciones o en iteradores. Este tipo de ciclo itera sobre una colección de objetos, en lugar de especificar valores o condiciones numéricos, este ciclo se llama ciclo for.

```python
for element in collection:
    # Block of code that iterates over the collection
```

Ejemplo en lenguaje C:

```c
for (int i=0; i < 100; i++) {
    // Ciclo
}
```

````python
#Ciclo con 10 iteraciones, en cada iteración se le sumará 1, y se imprimirá el valor.
a = 0
for i in range(10):
  a += 1
  print(a)
````

## **5.3 While**

Se encarga de ejecutar una misma acción “mientras que” una determinada
condición se cumpla:

````python
año = 2001
while año <= 2012:
    print("Informes del Año",año)
    año += 1
````

op.PNG

````python
n=1
while n<=50:
    print(n)
    n = 2*n+1
    
print("Programa terminado")
````
